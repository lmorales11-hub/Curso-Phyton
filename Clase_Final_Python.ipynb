```python
from ast import Return
#Crear una columna que contega el genero (masculino/femenino)
#Crear una función de payton que contenga las reglas
def asignar_genero():
  if "m" in valor:
    return "Masculino"
  else:
      return "femenino"
```

```python
def asignar_genero(valor):
  if "m" in str(valor):
    return "Masculino"
  elif "f" in str(valor):
    return "Femenino"
  else:
    return "No_definido" # Consider how to handle cases where neither 'm' nor 'f' is present

who2["genero"] = who2["variable"].apply(asignar_genero)
who2.head(100)
```

```python
def asignar_metodo(valor):
  if "rel" in valor:
    return "relapse"
  elif "sn" in valor:
    return "negative pulmonary smear"
  elif "sp" in valor:
    return "positive pulmonary smear"
  elif "ep" in valor:
    return "extrapulmonary"
  else:
    return "to define"
```

```python
who2["metodo"]=who2 ["variable"].apply(asignar_metodo)
who2.head()
```

```python
# age group (014 = 0-14 yrs of age, 1524 = 15-24, 2534 = 25-34, 3544 = 35-44 years of age, 4554 = 45-54, 5564 = 55-64, 65 = 65 years or older).
def asignar_grupoedad(valor):
  if "014" in valor:
    return "0-14"
  elif "1524" in valor:
    return "15-24"
  elif "2534" in valor:
    return "25-34"
  elif "3544" in valor:
    return "35-44"
  elif "4554" in valor:
    return "45-54"
  elif "5564" in valor:
    return "55-64"
  else:
    return "65+"
```

```python
who2["grupoedad"] = who2["variable"].apply(asignar_grupoedad)
who2.head()
```

```python
# remover la columna variable
who2.drop(columns=["variable"], inplace=True)
```

```python
# renombrar la columna value a casos
who2.rename(columns={"value":"cases"}, inplace=True)
```
